```python
import pandas as pd

df = pd.read_excel("clean_data.xlsx")
df.head()
```

```python
df1 = df.loc[df['txn_description']=='PAY/SALARY', :]
df1
```

```python
df1.count(axis=0)
```

```python
dummy_var = pd.get_dummies(df1['gender'])
df1 = pd.concat([df1, dummy_var],axis = 1)
df1.rename(columns={'F':'Female', 'M':'Male'}, inplace = True)
df1.drop('gender', axis = 1, inplace = True)
df1.head()
```

```python
df1.describe(include='all')
```

```python
ac_nums = df1['account'].unique()
```

```python
salary_list = []
age_list = []
balance_list = []
card_present_list = []
gender_list_f = []
gender_list_m = []
```

```python
for i in ac_nums:
    salary_list.append(df1.loc[df1['account']==i, 'amount'].sum())
    age_list.append(df1.loc[df1['account']==i, 'age'].mean())
    balance_list.append(df1.loc[df1['account']==i, 'balance'].mean())
    gender_list_f.append(df1.loc[df1['account']==i, 'Female'].value_counts().idxmax())
    gender_list_m.append(df1.loc[df1['account']==i, 'Male'].value_counts().idxmax())
    card_present_list.append(df1.loc[df1['account']==i, 'card_present_flag'].value_counts().idxmax())
```

```python
len(balance_list)
```

```python
data={'Ac_Num':ac_nums, 'Salary':salary_list, 'Age':age_list, 'Balance': balance_list, 'Card_present_flag':card_present_list,\
     'Female_yes':gender_list_f, 'Male_yes':gender_list_m}
df2 = pd.DataFrame(data)
df2
```

```python
df2[['Age', 'Balance', 'Salary', 'Card_present_flag', 'Female_yes', 'Male_yes']].corr()
```

```python
df2.drop('Card_present_flag', axis=1, inplace=True)
```

```python
df2
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
#x = df2[['Age']]
#y = df2['Salary']
x = df2[['Balance']]
y = df2['Salary']
#x = df2[['Female_yes']]
#y = df2['Salary']
#x = df2[['Male_yes']]
#y = df2['Salary']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

lr = LinearRegression()
lr.fit(x_train, y_train)
yhat = lr.predict(x_test)

yhat[0:5]
```

```python
lr.score(x_test, y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_train, y_train, color = 'Blue', label='Train data')
plt.scatter(x_test, y_test, color = 'Red', label='Test data')
plt.title('Balance vs Salary scatter plot')
plt.xlabel('Balance')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
def PolyPlot(model, independent, dependent):
    x1=np.linspace(0, 250000, 100)
    y1=model(x1)
    plt.plot(independent, dependent, '.', x1, y1, '-')
    plt.title('Balance vs Salary')
    plt.show()
    plt.close()
```

```python
PolyPlot(p, x1, y1)
```

```python
z = df2[['Age', 'Balance', 'Female_yes', 'Male_yes']]
y = df2['Salary']

z_train , z_test, y_train, y_test = train_test_split(z, y, test_size=0.3, random_state=0)
```

```python
poly = PolynomialFeatures(degree=4)

x_poly_train = poly.fit_transform(z_train)
x_poly_test = poly.fit_transform(z_test)

lr2 = LinearRegression()
lr2.fit(x_poly_train, y_train)
yhat2 = lr2.predict(x_poly_test)
yhat2[0:5]
```

```python
r_score2 = r2_score(y_test, yhat2)
r_score2
```

```python
ax3 = sns.distplot(y_test, hist=False, color='Blue', label='Actual Price')
ax4 = sns.distplot(yhat2, hist=False, color='Red', label='Fitted Price', ax=ax3)

plt.title('Salary vs Customer attributes')
plt.show()
plt.close()
```
