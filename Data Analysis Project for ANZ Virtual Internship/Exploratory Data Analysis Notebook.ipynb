```python
import pandas as pd

data= "https://insidesherpa.s3.amazonaws.com/vinternships/companyassets/AKkAyEwWc8wjPxx9n/Gz8nAM5drF9tcbBa4/ANZ%20synthesised%20transaction%20dataset.xlsx"
df=pd.read_excel(data)

df.head()
```

```python
df.info
```

```python
column_names=df.columns

for column in column_names:
    print("column name =", column, "\nnumber of missing values =", df[column].isnull().sum())
```

```python
import numpy as np
```

```python
card_present_flag_max_freq = df['card_present_flag'].value_counts().idxmax()
df['card_present_flag'].replace(np.nan, card_present_flag_max_freq, inplace=True)

merchant_id_max_freq = "106e1272-44ab-4dcb-a438-dd98e0071e51"
df['merchant_id'].replace(np.nan, merchant_id_max_freq, inplace=True)

merchant_suburb_max_freq = "Melbourne"
df['merchant_suburb'].replace(np.nan, merchant_suburb_max_freq, inplace=True)

merchant_state_max_freq = "NSW"
df['merchant_state'].replace(np.nan, merchant_state_max_freq, inplace=True)

merchant_long_lat_max_freq = "151.21 -33.87"
df['merchant_long_lat'].replace(np.nan, merchant_long_lat_max_freq, inplace=True)

df.drop(['bpay_biller_code', 'merchant_code'], axis=1, inplace=True)
```
